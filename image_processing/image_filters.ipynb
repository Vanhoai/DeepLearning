##### Blur
1. Gaussian Blur
2. Average Blur
3. Median Blur
4. Bilateral Filter

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread("image.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Gaussian Blur
gaussian_blur = cv2.GaussianBlur(img_rgb, (15, 15), 0)

# Average Blur
average_blur = cv2.blur(img_rgb, (15, 15))

# Median Blur
median_blur = cv2.medianBlur(img_rgb, 15)

# Hiển thị kết quả
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes[0,0].imshow(img_rgb)
axes[0,0].set_title('Original')
axes[0,1].imshow(gaussian_blur)
axes[0,1].set_title('Gaussian Blur')
axes[1,0].imshow(average_blur)
axes[1,0].set_title('Average Blur')
axes[1,1].imshow(median_blur)
axes[1,1].set_title('Median Blur')

plt.show()
```

##### Sharpen
1. Unsharp Masking
2. Kernel sharpening

```python
kernel_sharpen = np.array([[-1,-1,-1],
                          [-1, 9,-1],
                          [-1,-1,-1]])

# Áp dụng filter
sharpened = cv2.filter2D(img_rgb, -1, kernel_sharpen)

# Unsharp Masking - phương pháp sharpening nâng cao
gaussian = cv2.GaussianBlur(img_rgb, (0, 0), 2.0)
unsharp_mask = cv2.addWeighted(img_rgb, 1.5, gaussian, -0.5, 0)

# Hiển thị
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
axes[0].imshow(img_rgb)
axes[0].set_title('Original')
axes[1].imshow(sharpened)
axes[1].set_title('Kernel Sharpening')
axes[2].imshow(unsharp_mask)
axes[2].set_title('Unsharp Masking')
plt.show()
```

##### Edge Detection
1. Sobel
2. Canny
3. Laplacian

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Sobel Edge Detection
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = np.sqrt(sobelx**2 + sobely**2)

# Canny Edge Detection
canny = cv2.Canny(gray, 100, 200)

# Laplacian Edge Detection
laplacian = cv2.Laplacian(gray, cv2.CV_64F)

# Hiển thị
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes[0,0].imshow(gray, cmap='gray')
axes[0,0].set_title('Original Grayscale')
axes[0,1].imshow(sobelx, cmap='gray')
axes[0,1].set_title('Sobel X')
axes[0,2].imshow(sobely, cmap='gray')
axes[0,2].set_title('Sobel Y')
axes[1,0].imshow(sobel_combined, cmap='gray')
axes[1,0].set_title('Sobel Combined')
axes[1,1].imshow(canny, cmap='gray')
axes[1,1].set_title('Canny')
axes[1,2].imshow(laplacian, cmap='gray')
axes[1,2].set_title('Laplacian')
plt.show()
```
